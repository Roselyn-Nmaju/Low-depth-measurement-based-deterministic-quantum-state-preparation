```python
import numpy as np
import math as math
import matplotlib.pyplot as plt
import random as random
from qiskit.visualization import circuit_drawer
import DAC
```

Here we show the example of the "dense" input state from the paper

```python
input_vec = [np.sqrt(0.04), np.sqrt(0.13) ,np.sqrt(0.16) ,np.sqrt(0.2) ,np.sqrt(0.07) ,np.sqrt(0.09) ,np.sqrt(0.2) ,np.sqrt(0.11)]
circuit = DAC.DAC(input_vec,1)
circuit.draw("mpl", plot_barriers=False)
```

For the combine-and-conquer circuit we genereate a random imput vector that is not of length $2^n$ and also not normalised yet with simple modifications we can still encode this input using the divide-and-conquer circuit, then we show the structure of the circuit given differnt values for $\lambda$

```python
def GenRandom(x):
    rand = [random.random() for _ in range(x)]
    return rand
```

```python
input_vec_2 = GenRandom(14)
CAC_1 = DAC.DAC(input_vec_2,1)
CAC_2 = DAC.DAC(input_vec_2,2)
CAC_3 = DAC.DAC(input_vec_2,3)
CAC_4 = DAC.DAC(input_vec_2,4)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 20)) 

axs[0,0].set_title('=1')
circuit_drawer(CAC_1, output='mpl',plot_barriers=False, ax=axs[0,0])

axs[0,1].set_title('=2')
circuit_drawer(CAC_2,output='mpl',plot_barriers=False, ax=axs[0,1])

axs[1,0].set_title('=3')
circuit_drawer(CAC_3,output='mpl',plot_barriers=False,ax=axs[1,0])

axs[1,1].set_title('=4')
circuit_drawer(CAC_4,output='mpl',plot_barriers=False, ax=axs[1,1])

plt.tight_layout()
plt.show()
```
